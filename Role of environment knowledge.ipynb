```python
selected_columns = vco.iloc[:, 1:8]

result_df = pd.DataFrame()

for column in selected_columns.columns:
    column_counts = selected_columns[column].value_counts().reset_index()
    column_counts.columns = ['Value', 'Count']
    column_counts['Category'] = column
    column_counts = column_counts.sort_values(by='Value') 
    result_df = pd.concat([result_df, column_counts], axis=0)

result_df = result_df.reset_index(drop=True)

print(result_df)
```

```python
columns_to_compare = ['K1', 'K2', 'K3', 'K4', 'K5']

average_values = vco[columns_to_compare].mean()

plt.figure(figsize=(10, 6))
bars = plt.bar(average_values.index, average_values, color=['slategrey', 'royalblue', 'cornflowerblue', 'skyblue', 'lightsteelblue'])
plt.title('Comparison of K1, K2, K3, K4, and K5')
plt.xlabel('Knowledge')
plt.ylabel('Average Value')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black')

plt.show()
```

```python
sns.histplot(x= 'VCOWP', data = vco,color='skyblue')
for p in plt.gca().patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    plt.text(x + width/2, y + height, f'{height}', ha='center', va='bottom')

plt.show()
```

```python
vco_will_columns = ['VCOwill1', 'VCOwill2', 'VCOwill3', 'VCOwill4', 'VCOwill5']
vco_will_data = vco[vco_will_columns]

plt.figure(figsize=(10, 6))
ax = vco_will_data.apply(lambda x: x.value_counts()).T.plot(kind='bar', stacked=True)

for i, patch in enumerate(ax.patches):
    width, height = patch.get_width(), patch.get_height()
    x, y = patch.get_xy()
    ax.annotate(f'{height:.0f}', (x + width / 2, y + height / 2),
                ha='center', va='center', fontsize=8)

plt.title('Value Counts of VCOwill1 to VCOwill5')
plt.xlabel('VCOwill Columns')
plt.ylabel('Count')

ax.legend( bbox_to_anchor=(0.5, -0.25), loc='upper center', ncol=len(vco_will_columns))
plt.show()
```
